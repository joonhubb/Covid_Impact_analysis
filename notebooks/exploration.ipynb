```python
import pandas as pd

df = pd.read_csv(
    '../data/train_dataset/task_2-owid-covid-data-22_September_2020.csv'
)

df.head()
```

```python
global_death_rate = (
    df['total_deaths'].sum() / df['total_cases'].sum()
) * 100

global_death_rate
```

```python
country_death_rate = (
    df.groupby('location')['total_deaths'].max()
    / df.groupby('location')['total_cases'].max()
) * 100
country_death_rate.sort_values(ascending=False).head(10)
```

```python
df[['median_age', 'aged_65_older', 'aged_70_older']].describe()
```

```python
import matplotlib.pyplot as plt

filtered_df = df[df['total_cases'] > 10000]

country_death_rate_filtered = (
    filtered_df.groupby('location')['total_deaths'].max()
    / filtered_df.groupby('location')['total_cases'].max()
) * 100

country_death_rate_filtered.sort_values(ascending=False).head(10)
```

```python
import pandas as pd

def load_covid_data(path: str) -> pd.DataFrame:
    """
    Load OWID COVID dataset from given path
    """
    df = pd.read_csv(path)
    return df


def basic_cleaning(df: pd.DataFrame) -> pd.DataFrame:
    """
    Perform basic cleaning for analysis
    """
    cols_needed = [
        "location",
        "date",
        "total_deaths",
        "total_deaths_per_million",
        "people_vaccinated",
        "median_age",
        "aged_65_older"
    ]

    available_cols = [c for c in cols_needed if c in df.columns]
    df = df[available_cols]

    return df
```
